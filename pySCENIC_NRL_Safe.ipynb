## pySCENIC

```python
import pandas as pd
import numpy as np
import os, glob
import pickle

from arboreto.utils import load_tf_names
from arboreto.algo import grnboost2

from pyscenic.rnkdb import FeatherRankingDatabase as RankingDatabase
from pyscenic.utils import modules_from_adjacencies
from pyscenic.prune import prune2df, df2regulons
from pyscenic.aucell import aucell

import seaborn as sns
```

Paths to folders and files

```python
DATA_FOLDER="/mnt/dv/wid/projects5/Roy-singlecell/daniela_work/tmp"
RESOURCES_FOLDER="/mnt/dv/wid/projects5/Roy-singlecell/daniela_work/resources"
DATABASE_FOLDER = "/mnt/dv/wid/projects5/Roy-singlecell/daniela_work/databases/"

DATABASES_GLOB = os.path.join(DATABASE_FOLDER, "hg38*.mc9nr.feather")
MOTIF_ANNOTATIONS_FNAME = os.path.join(RESOURCES_FOLDER, "motifs-v9-nr.hgnc-m0.001-o0.0.tbl")

HS_TFS_FNAME = os.path.join(RESOURCES_FOLDER, 'hs_hgnc_tfs.txt')
SC_EXP_FNAME = os.path.join(RESOURCES_FOLDER, "NRL_Safe.tsv")

ADJACENCIES_FNAME = os.path.join(DATA_FOLDER, "adjacencies.tsv")
MODULES_FNAME = os.path.join(DATA_FOLDER, "modules.p")
MOTIFS_FNAME = os.path.join(DATA_FOLDER, "motifs.csv")
REGULONS_FNAME = os.path.join(DATA_FOLDER, "regulons.p")
```

Get expression data

```python
ex_matrix = pd.read_csv(SC_EXP_FNAME, sep='\t', header=0, index_col=0).T
```

```python
ex_matrix.head()
```

```python
ex_matrix.shape
```



List of known TFs for HS


```python
tf_names = load_tf_names(HS_TFS_FNAME)
```

Load ranking databases

```python
db_fnames = glob.glob(DATABASES_GLOB)
```

```python
def name(fname):
    return os.path.splitext(os.path.basename(fname))[0]
dbs = [RankingDatabase(fname=fname, name=name(fname)) for fname in db_fnames]
```

```python
dbs
```

Phase I 

Run GRNboost from arboreto to infer co-expression modules

```python
adjacencies = grnboost2(expression_data=ex_matrix, tf_names=tf_names, verbose=True)
```

```python
adjacencies.head()
```

```python
adjacencies.to_csv(ADJACENCIES_FNAME, index=False, sep='\t')
```

Derive potential regulomes from these co-expression modules

```python
modules = list(modules_from_adjacencies(adjacencies, ex_matrix))
```

```python
with open(MODULES_FNAME, 'wb') as f:
    pickle.dump(modules, f)
```


Phase II: Prune modules for targets with cis regulatory footprints (aka RcisTarget)¶


```python
df = prune2df(dbs, modules, MOTIF_ANNOTATIONS_FNAME)
```

```python
df.head()
```

```python
df.to_csv(MOTIFS_FNAME)
```



This dataframe can then be converted to regulons.


```python
regulons = df2regulons(df)
```

```python
with open(REGULONS_FNAME, 'wb') as f:
    pickle.dump(regulons, f)
```

Phase III: Cellular regulon enrichment matrix (aka AUCell)¶ 

```python
auc_mtx = aucell(ex_matrix, regulons, num_workers=1)
```

```python
sns.clustermap(auc_mtx, figsize=(12,12))
```
